```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
url = 'https://fbref.com/en/comps/9/history/Premier-League-Seasons'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')
```

```python
quotes = soup.find_all(scope="row", class_='left')
len(quotes)
```

```python
quotes[1]
```

```python
seasons = ['2017-2018','2018-2019', '2019-2020', '2020-2021']
```

```python
overall_stats_links = []
gca_stats_links = []
shooting_stats_links = []
possession_stats_links = []
for x in seasons:
    overall_stats_links.append(f'https://fbref.com//en/comps/9/{x}/{x}-Premier-League-Stats')
    gca_stats_links.append(f'https://fbref.com/en/comps/9/{x}/gca/{x}-Premier-League-Stats')
    shooting_stats_links.append(f'https://fbref.com/en/comps/9/{x}/shooting/{x}-Premier-League-Stats')
    possession_stats_links.append(f'https://fbref.com/en/comps/9/{x}/possession/{x}-Premier-League-Stats')
#all_links = [overall_stats_links, gca_stats_links, shooting_stats_links, possession_stats_links]
```

```python
# Created lists just to contain links #
overall_df = []
gca_df = []
shooting_df = []
possession_df = []
#__________________________________________________#
#Filling lists with right links
for link in overall_stats_links:
    overall_df.append(pd.read_html(link)[0])
```

```python
for link in gca_stats_links:
    gca_df.append(pd.read_html(link)[0])
for link in shooting_stats_links:
    shooting_df.append(pd.read_html(link)[0])
for link in possession_stats_links:
    possession_df.append(pd.read_html(link)[0])
```

```python

#__________________________________________________#

#deleting multiindex levels#
for season_data in possession_df:
    season_data.columns = season_data.columns.droplevel(0)
for season_data in gca_df:
    season_data.columns = season_data.columns.droplevel(0)
for season_data in shooting_df:
    season_data.columns = season_data.columns.droplevel(0)
```
